```
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import csv
```

```
headers = {
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:45.0) Gecko/20100101 Firefox/45.0'
      }
url = 'https://www.marvel.com/v1/pagination/grid_cards?offset={}&limit=36&entityType=character&sortField=title&sortDirection=asc'
links = []
for i in range(72):
    r = requests.get(url.format(i*36), headers = headers)
    a = json.loads(r.text)
    all_characters = a['data']['results']['data']
    for character in all_characters:
        links.append(character['link']['link'])
print(links)
```

```
print(len(links))
```

```
column_names = ['name', 'link', 'universe', 'other aliases', 'education', ' place of origin', 'identity', 'known relatives']
df = pd.DataFrame(columns = column_names)

base_url = 'https://www.marvel.com{}'
i = 0

for link in links:
  print('link '+str(i))
  character_info = []

  f = requests.get(base_url.format(link), headers = headers).text
  soup = BeautifulSoup(f)
  try:
    name = soup.find('span', {'class': 'masthead__headline'}).text
  except AttributeError:
      print('404 not found')
  character_info.append(name)
  character_info.append('www.marvel.com' + link)
  rail_bio_links = soup.find_all('ul', {'class': 'railBioLinks'})
  for rail_bio_link in rail_bio_links:
    character_info.append(rail_bio_link.text)
  if(len(character_info)==8):
    df = df.append(pd.Series(character_info, index=df.columns), ignore_index=True)
  i+=1
```

```
df.to_csv('marvel.csv')
```
